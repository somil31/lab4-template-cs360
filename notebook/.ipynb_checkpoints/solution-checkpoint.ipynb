```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# ===== Optional : import other libraries here ===== #

# ===== End of Optional : import other libraries here ===== #
```

## Problem 1 (a): Download dataset
According to instruction PDF, download [Breast Cancer Coimbra Data Set](https://archive.ics.uci.edu/dataset/451/breast+cancer+coimbra) and extract file `dataR2.csv` to directory `./data`.

```python
# Visualization: Load dataset

df = pd.read_csv('data/dataR2.csv')
features_pd = df.copy()
classification = features_pd.pop('Classification').to_numpy().astype(int)
print(">>> example features:\n", features_pd.head(5))
print(">>> example classification:", classification[:5])
print(">>> classification has unique (labels, counts):", np.unique(classification, return_counts=True))
features = features_pd.to_numpy().astype(np.float32)
print(">>> features has shape:", features.shape)
```

## Problem 1 (b): Exploratory data analysis
- Convert `classification` from [1, 2] to [0, 1] and save in new numpy array `classification_converted`.
- Plot scatter matrix with class=0 being red and class=1 being green.
- Select the first 40 rows of Class 0 and the first 48 rows of Class 1 as the test set (`training_data`, `test_class`) and the rest of the data as the test set (`test_data`, `test_class`).

```python
classification_converted = classification.copy()
test_data, training_class = None, None
test_data, test_class = None, None

# ===== Convert classification to 0 or 1 ===== #
mask = np.where(classification == 1, True, False)
classification_converted[mask] = 0
classification_converted[~mask] = 1
# ===== End of Convert classification to 0 or 1 ===== #

# ===== Make a colored scatter matrix ===== #
# Hint: use variable features_pd and pd.plotting.scatter_matrix()
colors = np.zeros((len(classification_converted), 3))
colors[mask] = [1., 0., 0.]
colors[~mask] = [0., 1., 0.]
pd.plotting.scatter_matrix(features_pd, alpha=0.75, figsize=(10, 10), c=colors)
plt.show()
# ===== End of Make a colored scatter matrix ===== #

# ===== Divide into training and test set ===== #
training_data = np.vstack([features[np.where(classification_converted == 0)][:40], features[np.where(classification_converted == 1)][:48]])
training_class = np.hstack([classification_converted[np.where(classification_converted == 0)][:40], classification_converted[np.where(classification_converted == 1)][:48]])
test_data = np.vstack([features[np.where(classification_converted == 0)][40:], features[np.where(classification_converted == 1)][48:]])
test_class = np.hstack([classification_converted[np.where(classification_converted == 0)][40:], classification_converted[np.where(classification_converted == 1)][48:]])
# ===== End of Divide into training and test set ===== #
```

```python
# TEST: 1(b)
assert(len(np.where(classification_converted == 0)[0]) == 52)
assert(len(np.where(classification_converted == 1)[0]) == 64)
assert(len(np.unique(classification_converted)) == 2)

assert(training_data.shape[0] == 88 and training_data.shape[1] == 9)
assert(test_data.shape[0] == 28 and test_data.shape[1] == 9)
assert(training_class.shape[0] == 88)
assert(test_class.shape[0] == 28)

print(f">>> Passed P1(b) Tests.")
```

## Problem 1 (c): Classification using KNN on Breast Cancer Coimbra Data Set
- Train kNN module with training data. Set `weights` to `uniform`, `metric` to `euclidean`.
- Compute misclassification rates on both training and test dataset.
- Find and report the best k.

```python
# ===== Classify training/test data ===== #
training_misclassification_rates, test_misclassification_rates = [], []
for k in reversed(range(1, 89)):
    nbrs = KNeighborsClassifier(n_neighbors=k, algorithm='ball_tree', weights='uniform', metric='euclidean').fit(training_data, training_class)
    training_result = nbrs.predict(training_data)
    test_result = nbrs.predict(test_data)
    training_misclassification_rates.append(np.sum(training_class != training_result) / training_class.shape[0] * 100.)
    test_misclassification_rates.append(np.sum(test_class != test_result) / test_class.shape[0] * 100.)
plt.plot(89-(np.arange(88)+1), training_misclassification_rates, label="Training")
plt.plot(89-(np.arange(88)+1), test_misclassification_rates, label="Test")
# ===== End of Classify training/test data ===== #

# ===== Plot misclassification rate ===== #
plt.xlabel("$k$")
plt.ylabel("Misclassification Rate (%)")
plt.title("Misclassification Rates on Training/Test Data")
plt.xlim(88, 0)
plt.legend()
plt.show()
# ===== End of Plot misclassification rate ===== #
```

```python
# ===== Which k^* is the most suitable k ===== #
k_star = (89-(np.arange(88)+1))[np.array(test_misclassification_rates).argmin()]
print(f"k_star = {k_star}")
# ANSWER: K^* is {46}.
# REASON: Misclassification rate on test set is the smallest.
# ===== End of Which k^* is the most suitable k ===== #
```

## Problem 1 (d): Replace the Euclidean metric with the Minkowski distance
- Train kNN module with training data. Set `weights` to `uniform`, `metric` to `minkowski`.
- Compute misclassification rates on both training and test dataset.
- Find and report the best k and p.

```python
# ===== Classify training/test data ===== #
training_misclassification_rates, test_misclassification_rates = [], []
for k in reversed(range(1, 89)):
    training_misclassification_rates_p, test_misclassification_rates_p = [], []
    for p in range(1, 6):
        nbrs = KNeighborsClassifier(n_neighbors=k, algorithm='ball_tree', weights='uniform', metric='minkowski').fit(training_data, training_class)
        training_result = nbrs.predict(training_data)
        test_result = nbrs.predict(test_data)
        training_misclassification_rates_p.append(np.sum(training_class != training_result) / training_class.shape[0] * 100.)
        test_misclassification_rates_p.append(np.sum(test_class != test_result) / test_class.shape[0] * 100.)
    training_misclassification_rates.append(training_misclassification_rates_p)
    test_misclassification_rates.append(test_misclassification_rates_p)
training_misclassification_rates = np.array(training_misclassification_rates)
test_misclassification_rates = np.array(test_misclassification_rates)
# ===== End of Classify training/test data ===== #

# ===== Find the optimal (k^*, p^*) ===== #
index = np.unravel_index(np.argmin(test_misclassification_rates), test_misclassification_rates.shape)
k_star = (89-(np.arange(88)+1))[index[0]]
p_star = (range(1, 6))[index[1]]
print(">>> optimal k is", k_star, ", optimal p is", p_star)
# ANSWER: k^* is {46}, p^* is {1}.
# ===== End of Find the optimal (k, p) ===== #
```
